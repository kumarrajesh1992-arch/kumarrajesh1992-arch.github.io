```python
import pandas as pd
df = pd.read_csv("PP434_Project_Data_1.csv")
df.columns
```

```python
df.info()
df.isna().sum()
```

```python
def hdi_category(x):
    if x >= 0.8:
        return "Very High"
    elif x >= 0.7:
        return "High"
    elif x >= 0.55:
        return "Medium"
    else:
        return "Low"

df["HDI_category"] = df["SHDI"].apply(hdi_category)
```
